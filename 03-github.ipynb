## 03. GitHub

### git and GitHub

git은 버전 관리 시스템이고, [GitHub](https://github.com)는 git으로 관리하는 프로젝트를 올려둘 수 있는 git 원격 저장소를 지원하는 웹 사이트입니다. git은 파일의 변경 사항에 대해 언제, 누가, 무엇을, 어떻게 변경했는지를 비교할 수 있는 스냅샷 기반의 분산 버전 관리 시스템으로 프로그램 개발 과정에서 소스 코드를 관리하고 협업을 도와주는 도구입니다. GitHub는 git으로 관리하는 프로젝트를 올리고 공유할 수 있는 대표적인 git 원격 저장소 호스팅 서비스 중 하나입니다. 다른 호스팅 사이트로는 [Bitbucket](https://bitbucket.org/), [GitLab](https://about.gitlab.com/)  등이 있습니다. GitHub은 이러한 호스팅 사이트 중 거의 표준에 가까운 서비스입니다. git은 컴퓨터 파일 시스템을 추적 대상으로 하기 때문에 파일 시스템이 구성된 장치라면 어디에서나 사용할 수 있습니다. 예를 들어, USB에 .git 디렉토리가 있다면, USB를 연결한 사람의 컴퓨터에서 git을 사용할 수 있습니다. 따라서, git 기능이 포함된 웹 호스팅 서버가 존재한다면, 웹을 통해 여러 사람들과 함께 협업이 가능합니다.

git을 비롯하여 버전이과 백업을 관리해주는 Dropbox, Onedrive, Google Drive와 같은 클라우드 서비스들은 모두 클라이언트-서버 구조를 가지고 있습니다. 여기서 클라이언트는 우리가 각 웹 사이트에서 다운로드받는 프로그램 입니다. 즉, 서비스의 사용자는 클라이언트-서버 구조 중 클라이언트에 해당합니다.

예를 들어 Onedrive의 클라이언트로써 우리가 프로그램을 사용하는 방식을 생각해보면, 우리는 간단히 파일을 만들고 저장하고, 단지 그것만 수행합니다. 그러면, 서버는 어떤 일을 할까요? 서버는 사용자가 저장한 파일을 서버에 입력받아, **가능한 실시간에 가깝게, 가능한 충돌이 발생하지 않게 저장 및 버저닝**한 후, 파일을 사용하는 사용자가 해당 파일을 요청하면 파일을 전송합니다.

git의 관점으로 바라보면, 우리의 로컬 git 저장소는 하나의 클라이언트,  GitHub는 서버의 역할을 합니다. 우리가 로컬 git 저장소에서 `git add`, `git commit`과 같은 명령어들을 이용하여 작업한 결과물들을 `git push` 명령어를 이용하여 GitHub 로 전송합니다. GitHub 서버는 우리가 전송한 git 저장소를 가능한 실시간에 가깝게, 가능한 충돌이 발생하지 않게 저장합니다. 그리고 저장된 git 저장소를 웹에 공개합니다. 본 부트캠프의 강의 자료들이 올라가있는 웹 사이트가 바로 GitHub입니다.

우리가 지난 git 수업시간에서 다룬 내용을 생각해보면, 로컬의 git 저장소만 이용했습니다. 즉, 아직 다른 사람들과 공유하는 저장소를 다뤄보지 못했습니다. 이번 시간에는 다른 사람들과 공유할 수 있는 GitHub의 원격 저장소 기능을 활용해볼 것입니다.

### GitHub for Career

단순하게는 GitHub은 git의 원격 저장소로써 활용되고, 소스 코드 혹은 어떠한 자료들을 웹 상에 공개하는 것으로 생각할 수 있습니다. 그런데, 이러한 이유로 어떠한 시점부터 많은 소프트웨어 회사들은 채용 시에 지원자의 GitHub 링크를 작성하는 것을 적극 권장하고, 일부 스타트업들은 GitHub 링크 제출을 의무화하고 있습니다.

GitHub는 개발자, 분석가, 디자이너 등의 이력서 및 포트폴리오를 상당수 대체해주는 역할을 하고 있습니다. 지원자가 1) *어떤 프로젝트를 진행해왔는지*, 2) *git을 잘 활용할 줄 아는지*, 3) *협업 경험이 있는지* 등 GitHub에는 많은 나의 정보들을 기록해두고 이를 채용에 활용할 수 있습니다. 또한 GitHub는 *github.io* 라는 웹 호스팅 서비스도 지원하고 있어, 이를 활용하여 간단한 블로그 혹은 개인 웹 사이트를 만드는 것도 가능합니다.

GitHub를 활용한 [이력서의 좋은 예시](), github.io를 활용한 [하이퍼커넥트 기술 블로그](https://hyperconnect.github.io/)

연구자에게도 GitHub는 강력한 기회들을 제공합니다. 예를 들면, GitHub를 통해 연구를 쉽게 공유하고, 공유받을 수 있으며 공개된 연구의 내용대로 잘 동작하는지 직접 테스트하는 것도 가능합니다. git으로 잘 관리된 프로젝트의 경우 커밋 메세지를 통해 연구 과정과 중간 결과물들도 확인할 수 있습니다. GitHub을 통해 진행되고 있는 오픈 소스 프로젝트에 기여하는 것도 가능합니다. 잘 알려진 오픈 소스 프로젝트에 기여했다는 것은 해당 오픈 소스를 잘 활용하고 있고, 문제점 혹은 개선점을 발견하여 해결할 수 있는 기술적 능력이 있다는 이야기므로, 많은 경우에 좋은 커리어로써의 역할을 합니다. 기술적인 기여가 아니더라도, 해당 기술들을 다른 언어로 번역하거나 (영어 $\leftrightarrow$ 한국어, Python2 $\leftrightarrow$ Python3), 오타를 수정하는 것도 오픈소스에 기여하는 좋은 자세입니다.

또한 꼭 코드를 기여해야 하는 것은 아닙니다. [과제물들을 잘 정리](https://github.com/mybirth0407/graduate_courses)해둘 수도 있을 것이고, 수업 필기를 업로드하고 공유하는 것도 GitHub를 활용한 커리어 쌓기에 해당됩니다.

### git and Github for Collaboration

#### Local and Remote Repository

GitHub에 프로젝트를 올리고 공개 저장소로 설정한다면, 시간, 공간적 제한 없이 전 세계 엔지니어들과 협업할 수 있습니다. 아마 통계학도라면 한번 쯤은 경험해봤을 [RStudio](https://github.com/rstudio/) 또한 GitHub에서 공유되고 있는 오픈 소스 프로젝트입니다. 그 외에도 GitHub를 통해 오픈 소스로 개발되고 있는 프로젝트는 수없이 많습니다. GitHub에 자료를 올려 공개하는 것만으로도 우리는 오픈 소스 활동을 하고 있는 것입니다.

그러면, GitHub를 이용한 첫 번째 원격 저장소를 만들어보도록 하겠습니다. GitHub의 원격 저장소와, git 로컬 저장소를 연결시키는 방법은 두 가지가 있습니다.

1. GitHub 원격 저장소를 먼저 생성한 후, 로컬 저장소로 `clone` 합니다. 이 방법은 간편하지만, 인터넷에 연결되어 있어야 하고, 프로젝트 시작 첫 단계에 이루어져야 합니다.
3. GitHub 원격 저장소를 생성합니다. 그리고 로컬 저장소에서 작업하던 내용을 GitHub 원격 저장소에 `push` 합니다. 이 방법은 인터넷에 연결되지 않은 상태여도 프로젝트를 먼저 진행할 수 있습니다.

두 가지 방법을 모두 다뤄보고, 각각의 장단점을 체감해보도록 하겠습니다.

먼저, 원격 저장소를 생성해보도록 하겠습니다.

1. https://github.com 에 접속합니다.
2. 왼쪽 상단의 `New` 버튼을 클릭해 GitHub에 새 원격 저장소를 생성합니다.
3. Repository name에 `hello-github` 라고 입력합니다. 다른 이름을 입력해도 괜찮습니다. GitHub의 저장소명은 다른 저장소와는 중복되어도 괜찮지만 **내 소유의 저장소와는 중복되면 안됩니다.**
4. `Description`은 git 저장소가 추적하는 파일이 아닌, GitHub에서만 보이는 간단한 프로젝트 설명입니다. `Fun with GitHub` 이라고 입력합니다. 비워두셔도 괜찮습니다.
5. `Add a README file` 옆의 체크박스를 체크합니다. README 파일은 주로 마크다운 (.md, markdown foramt)으로 작성되며, 프로젝트의 간단한 소개와 사용법 등을 기록합니다. README 파일은 git 저장소가 추적하는 파일입니다.
6. `Create repository` 버튼을 클릭합니다.

다음으로, 터미널 환경으로 돌아와서 로컬 환경에 원격 저장소를 가져오도록 하겠습니다. 먼저 로컬 환경에서 오늘 실습을 진행할 `github` 디렉토리를 생성하고 `github` 디렉토리로 이동합니다.
```bash
mkdir github
cd github
```

`github` 디렉토리에서 `git clone` 명령어를 이용하여 원격 저장소를 가져옵니다. `git clone` 명령어는 원격 저장소를 로컬 환경으로 복제하는 명령어입니다. `git clone` 명령어는 원격 저장소의 주소를 인자로 받습니다. 원격 저장소의 주소는 웹 브라우저에서 해당 저장소를 방문하면 주소창에 표시됩니다. `git clone` 명령어를 실행하면, 원격 저장소의 내용을 복제하여 `hello-github` 로컬 저장소를 생성합니다.
```bash
git clone https://github.com/[ID]/hello-github
```

`ls` 명령어를 통해 github 디렉토리 내에 hello-github 디렉토리가 다운로드 된 것을 확인할 수 있습니다. `hello-github` 디렉토리로 이동한 후, 원격 저장소의 README.md 파일이 존재하는지, 그리고 같은 내용이 출력되는지 확인합니다.
```bash
cd hello-github
cat README.md
```

`# hello-github`<br>`Fun with Github`

메세지가 출력되면 정상적으로 출력된 것입니다. 따라서, 우리는 현재 원격 저장소의 내용을 로컬 저장소로 복제하였습니다. 이제 로컬 저장소에서 작업을 하고, 그 결과물을 원격 저장소에 업로드해보도록 하겠습니다.

touch 명령어를 이용하여 `hello.txt` 파일을 생성하고 `Hello, GitHub` 라는 내용을 작성합니다. 파일 생성 후 `git add` 명령어를 이용하여 `staging area`로 이동시킵니다. `staging area` 의 파일은 commit 되기 전까지 로컬 저장소에만 기록되는 파일입니다. `git status` 명령어를 실행하면 `hello.txt` 파일이 `staging area`로 이동한 것을 확인할 수 있습니다.

```bash
touch hello.txt
echo "Hello, GitHub" > hello.txt
git add hello.txt
git status
```

`staging area`의 파일을 커밋하여 로컬 저장소에 기록합니다. 메세지는 `add hello.txt` 혹은 다른 메세지로 작성해도 좋습니다. 커밋 후에도 원격 저장소에는 아무 변화가 없는 것을 확인합니다.
```bash
git commit -m "add hello.txt"
```

로컬 저장소에서 커밋된 내용을 원격 저장소로 푸시 (push) 해야 합니다. `git push` 명령어를 통해 원격 저장소로 푸시할 수 있습니다. 푸시할 때에는 원격 저장소의 명칭과 브랜치를 정확하게 입력해주는 것이 좋습니다. 원격 저장소의 기본 이름은 `origin` 이고, 기본 브랜치명은 `main` 입니다.

```bash
git push origin main
```

원격 저장소의 변경 사항을 직접 확인합니다.

원격 저장소에 `hello.txt` 라는 파일이 생성되었고, 그 내용은 `hello github` 인 것을 확인할 수 있습니다. 다음으로는 원격 저장소에서 파일을 추가한 후, 로컬 저장소로 가져오는 방법을 알아보도록 하겠습니다.

1. 원격 저장소의 `hello.txt` 파일을 클릭합니다.
2. 우측 중상단의 연필 모양 아이콘 (Edit this file)을 클릭하여 텍스트 편집기 화면으로 이동합니다.
3. 텍스트 편집기 화면에서 `Hello, GitHub` 를 `Hello, GitHub!` 로 수정합니다. 다르게 변경하셔도 괜찮습니다.
4. 우측 상단의 `Commit changes` 버튼을 클릭하여 커밋 메세지를 작성합니다. 기본 커밋 메세지는 `Update [파일명]` 입니다. 다르게 작성하셔도 괜찮습니다.
5. `Extended description` 에는 커밋 메세지에 대한 설명을 작성합니다. `My first commit in github` 이라고 작성합니다. 비워두셔도 괜찮습니다.
6. `Commit changes` 버튼을 클릭하여 원격 저장소에 커밋 내역을 반영합니다.

원격 저장소의 경우 푸시해주지 않아도 됩니다. 푸시는 로컬 저장소의 변경사항을 원격 저장소로 업로드하는 것인데, 원격 저장소의 관점에서는 커밋만 발생하더라도 이미 변경사항이 원격 저장소에 반영된 것이기 때문입니다.

원격 저장소의 변경 사항을 로컬 저장소로 가져오는 방법은 두 가지가 있습니다.

1. `git pull` 기능을 이용하여 원격 저장소의 변경 사항을 웹 페이지에서 확인한 후 변경 사항을 로컬 저장소에 바로 반영합니다.
2. `git fetch` 원격 저장소의 변경 사항만을 로컬 저장소로 가져온 후, 변경 사항을 로컬 환경에서 확인합니다. 그 후 `git pull` 명령어를 통해 로컬 저장소에 반영합니다.

먼저 첫번째 방식을 활용해서 로컬 저장소에 변경 사항을 반영해보도록 하겠습니다. 원격 저장소 페이지의 중간 즈음의 `commits` 버튼을 클릭합니다. 버튼을 클릭하면 현재 저장소, 지정된 브랜치에 해당하는 커밋 내역이 웹 화면에 보여집니다. `Extended description`를 작성했다면 커밋 메세지 뒤에 $\dots$ 기호가 출력됩니다. 해당 기호를 클릭하면 작성한 `Extended description`를 확인할 수 있습니다. 커밋 메세지를 눌러 변경 사항을 확인합니다. 변경 사항을 확인했다면, 터미널 환경으로 돌아옵니다.

GitHub에서 원격 저장소의 변경 사항을 확인했으므로, 로컬 저장소에 반영해보도록 하겠습니다. 방법은 간단합니다. `git pull` 명령어를 이용합니다. `git push` 명령어와 마찬가지로 원격 저장소의 이름과 브랜치 이름을 명시해주는 것이 좋습니다.

```bash
git pull origin main
```

원격 저장소에서 발생한 변경사항이 잘 적용되었는지 확인합니다.
```bash
cat hello.txt
```

다음으로 `git fetch`를 이용해서 원격 저장소의 변경 사항만을 가져와보도록 하겠습니다. 먼저 변경 사항을 만들기 위해, GitHub 원격 저장소에서 파일을 생성합니다. 중간부의 `Add file` 버튼을 클릭한 후 `Create new file`을 클릭합니다. 왼쪽 상단의 `Name your file` 에 `world.txt` 라고 입력합니다. 그리고 `Hello, World!` 라고 입력합니다. 커밋 메세지를 작성하고 커밋합니다.

터미널 환경으로 돌아온 후에 로컬 저장소에 원격 저장소의 **변경 사항**만을 가져옵니다. `git status` 명령어를 입력하면, 로컬 저장소가 원격 저장소보다 얼마나 옛날 버전인지 알 수 있습니다. 지금은 한 단계만 차이날 것입니다.
```bash
git fetch origin main
git status
```

다음으로, 변경 사항이 어떤 내역인지 로컬 환경에서 확인합니다.
```bash
git diff ...origin
```

변경 사항을 확인한 후 합리적인 변경 사항이라면, `git merge` 명령어를 이용해 변경사항을 로컬 저장소에 병합합니다.
```bash
git merge
```

원격 저장소에서 발생한 변경사항이 잘 적용되었는지 확인합니다.
```bash
cat hello.txt
```

따라서, `git pull` 명령어는 `git fetch` 명령어와 `git merge` 명령어 두 개를 합쳐놓은 것과 같습니다. 가능하다면, `git fetch` 명령어를 이용해 변경 사항을 로컬 환경에서 충분히 검사한 후, `git merge`를 수행하는 것이 좋습니다.

지금까지는 git과 GitHub 모두 혼자 사용하고 있는 저장소에 대해서만 실습했습니다. 즉, 아직 협업 상황에서의 git과 GitHub에 대해서는 다뤄보지 않았습니다. 많은 경우 협업 과정에서 문제들이 발생합니다. 따라서 이를 방지하는 것, 그리고 이를 잘 해결하는 것이 중요합니다. 대부분의 경우 git은 문제를 해결하는 데 도움을 줍니다. 이것이 git이 단지 버전 관리 시스템의 역할이 아니라 훌륭한 협업 도구로써 활용되는 이유입니다.

#### Collaboration with Others

git 및 GitHub를 잘 활용하다는 것은, 각각의 내용을 잘 이해하는 것도 있겠지만, 두 가지 기술을 적절히 활용하여 문제 없이 협업하는 것을 의미합니다. 협업 과정에서는 다양한 문제가 발생합니다. 같은 파일을 여러 사람이 동시에 수정하거나, 특정 사람이 저장소를 망가뜨리는 경우 등이 있습니다.

우리는 한 명이지만, 두 명인 것처럼 실습을 해보도록 하겠습니다. **A**는 터미널 환경에서의 로컬 저장소에서 작업하고, **B**는 웹 환경에서의 원격 저장소에서 작업하도록 하겠습니다. 먼저, 터미널 환경에서 작업하는 **A**가 되어보겠습니다. 로컬 환경에서 작업하는 **A**, 현재 `hello.txt` 의 인사 문구가 너무 짧아 마음에 들지 않습니다. 그래서 파일에 다음과 같이 자기 소개를 추가하기로 결정했습니다.

```bash
echo "I am a Ph.D Student in Statistical computing Lab." >> hello.txt
```

문구가 마음에 들었던 **A**는 변경 사항을 커밋해둔 후 다른 작업을 하러 갔습니다.
```bash
git add hello.txt
git commit -m "add a my bio to hello.txt"
```

**A**는 다른 할 일을 하러 갔으니 우리는 **B**가 되어봅니다. **B**는 GitHub 페이지에서 파일을 검색하던 도중, `hello.txt`을 보고 GitHub에게만 인사하는 것이 미안해서, git 에게도 인사를 해주고 싶었습니다.. GitHub 페이지에서 파일 클릭 후, `Hello, git!`을 추가한 다음 커밋을 진행했습니다.

다시 **A**로 돌아가봅니다. **A**는 문득 인사 문구를 GitHub 원격 저장소로 푸시하지 않은 것이 생각났습니다. 터미널 환경으로 돌아와 푸시를 합니다.
```bash
git push origin main
```

그러나, **A**는 다음과 같은 메세지를 받았습니다.

당황한 **A**는 GitHub 원격 저장소를 들어가 확인했는데, `hello.txt`가 수정된 것을 확인했습니다. **A**는 어떻게 해결할 수 있을까요?

먼저, 다음과 같은 상황을 가정합니다. 로컬 git 저장소는 **D** 시점의 원격 저장소의 내용을 가져온 후, **E, F** 커밋을 거쳐 **G** 커밋까지 진행된 상태입니다. 원격 git 저장소는 **A, B** 커밋을 거쳐 **C** 커밋까지 진행되었습니다.

`pull.rebase false` 옵션을 이용하면 다음과 같이 두 브랜치가 병합되며, 로컬 git 저장소에 새로운 커밋을 자동으로 생성합니다. **Default** 설정입니다.



`pull.rebase true` 옵션을 사용하면 나뉘어진 브랜치들이 다음과 같이 병합됩니다.



`pull.ff only` 옵션은 브랜치의 그래프가 `fast-forward` 형태일 때만 가능합니다. 즉 특정 조건을 만족하는 상황에서만 사용이 가능한 병합입니다.

여기서는 **Default** 설정인 `pull.rebase false` 옵션을 이용하여 `git pull`을 진행합니다. 앞으로 모든 풀 명령어를 `pull.rebase false`로 진행하기 위해 다음과 같이 설정합니다.

```bash
git config --global pull.rebase false
```

다시 풀 명령어를 통해 변경사항을 가져옵니다.

```bash
git pull origin main
```

같은 파일에 대해 서로 다른 내용이 수정되어 *충돌 (conflict)*이 발생했습니다. 이렇게 충돌이 발생한 경우 파일에 직접 접근하여 직접 수정을 해주어야 합니다. 파일을 열어 확인을 해봅니다.

파일을 열면, 1) `<<<<<<< HEAD`, `=======` 라는 행과 2) `>>>>>>> 555982578bb9b041b3f7ade5e3ffb92cc293a68b` 라는 행이 추가되었습니다. 2)는 커밋 이름이므로 다를 수 있습니다. 여기서 우리가 살펴볼 구간은, `<<<<<<< HEAD` 이하와 `=======` 이상인 부분과 `>>>>>>> 555982578bb9b041b3f7ade5e3ffb92cc293a68b` 이하 부분입니다. 먼저, `<<<<<<< HEAD` 이하는 로컬 git 저장소에 기록된 최신 버전입니다. 반면, `>>>>>>> 555982578bb9b041b3f7ade5e3ffb92cc293a68b` 이하는 원격 저장소에 반영된 최신 버전입니다. 우리는 이 파일이 두 경로 이상으로부터 수정된 것을 알고 있고, 이 수정사항을 적절히 반영하여 로컬 git 저장소와 원격 git 저장소에 기록하는 것이 중요합니다.

```
Hello, GitHub
Hello, git!
I am a Phd student.
```
